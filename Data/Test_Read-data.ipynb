```python
import os
import numpy as np
import pandas as pd
import pickle # for serializing and de-serializing a Python object structure.
from datetime import datetime

import quandl

import plotly.offline as py
import plotly.graph_objs as go
import plotly.figure_factory as ff
py.init_notebook_mode(connected=True)
```

```python
def plotscatter(df, name='Stock name', title='Stock price', column='Close'):
    # 1) Trace
    df_trace = go.Scatter(x=df.index, y=df[column], name = name)
    
    # 2) Dados
    df_data = [df_trace]

    # 3) Layout
    df_layout = go.Layout(title=title, legend={'orientation':'h'}, yaxis={'title':'Preco'})

    # 4) Figure
    df_fig = go.Figure(data=df_data, layout=df_layout)
    py.iplot(df_fig)
```

```python
datasets = ['coffee', 'gold', 'sp500_fut', 'ibov', 'nasdaq', 'dji', 'goog', 'amzn', 'aapl', 'tsla', 'twtr', 
            'fb', 'ge', 'msft', 'real_usd', 'real_peso']
for ds in datasets:
    with open('{}.pkl'.format(ds), 'rb') as f:
        df = pickle.load(f)
        plotscatter(df, title=ds)
```
